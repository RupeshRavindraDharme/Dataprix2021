```python
from sklearn.preprocessing import LabelEncoder
df1 = df.fillna(df.mode().iloc[0]).iloc[:,:-1]
df1['Metabolic Syndrome Type']=df['Metabolic Syndrome Type']
df1
```

```python
df1.iloc[:,:-1] = df1.iloc[:,:-1].apply(LabelEncoder().fit_transform)
df1
```

```python
import seaborn as sns
corr = df1.corr()
sns.heatmap(corr, cmap='coolwarm')
```

```python
df1.head()
```

```python
X = df1.drop('Metabolic Syndrome Type', axis = 1)
y = df1['Metabolic Syndrome Type']
X.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
Xtrain.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(Xtrain, ytrain)
model.score(Xtest, ytest)
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
X2 = test.apply(LabelEncoder().fit_transform)
X2
```

```python
pred = model.predict(X2)
len(pred)
```
